```python
#importing the basic library
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

y=x*2

```python
y
```

```python
plt.plot(x,y)
```

Line chart

```python
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

Bar chart

```python
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.bar(x,y, color='pink')
plt.show()
```

scatter Plot

```python
a=(4,2,5,5,7,2,7,9,8)
b=(44,33,55,69,32,12,45,56,45)

plt.scatter(a,b)
plt.title("scatter Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.scatter(a,b,color='pink')
plt.show()
```

Histogram

```python
H=(3,3,4,3,5,6,4,4,4,4,4,4,4,7,6,7,9,6,5,4,3,9,9,9)
```

```python
plt.hist(H)
plt.hist(H,color='pink')
plt.show()
```
